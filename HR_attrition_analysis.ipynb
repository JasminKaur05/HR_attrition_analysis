```python
plt.figure(figsize=(7,5))
sns.histplot(df['Salary'],bins=30,kde=True)
plt.title("Salary Distribution")
plt.show()
```

The salary distribution is right-skewed, indicating that most employees earn moderate salaries while a small group receives significantly higher pay.

```python
plt.figure(figsize=(7,5))
sns.histplot(df['EmpSatisfaction'],bins=10,kde=True)
plt.title("Employee Satisfaction Distribution")
plt.show()
```

Most employees report moderate to high satisfaction levels, while a small group with low satisfaction may contribute to higher attrition risk.

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Termd', data=df)
plt.title("Attrition Distribution")
plt.xlabel("Attrition (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

The plot shows that the majority of employees have not left the organization, but a noticeable portion has churned, indicating a non-negligible attrition risk that requires attention.


```python
plt.figure(figsize=(8,5))
sns.countplot(x='Department',hue = 'Termd',data=df)
plt.title("Attrition by Department")
plt.xticks(rotation=45)
plt.show()
```

Attrition is highest in the Production department, suggesting heavier workload or job stress compared to other departments, while executive and admin roles show minimal turnover.


```python
plt.figure(figsize=(10,5))
sns.countplot(x='Position',hue='Termd',data=df)
plt.title("Attrition by Job Role")
plt.xticks(rotation=90)
plt.show()
```

Certain operational and entry-level roles experience higher attrition, whereas senior and specialized positions show stronger employee retention, indicating stability increases with role seniority.


## Final Conclusion

This analysis highlights that employee attrition is influenced more by organizational and role-based factors than by individual demographics.  
Departments with operational workloads and certain job roles experience higher attrition, while senior and specialized roles show stronger retention.  
The insights from this analysis can help HR teams focus on targeted retention strategies, workload balance, and role-specific engagement initiatives to reduce employee turnover.
